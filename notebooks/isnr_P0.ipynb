```python
import numpy as np
import matplotlib.pyplot as plt
```

# Figure 4

```python
label_font = 8.5
legend_font = 8
```

```python
klin = np.load(path_to_repo+"data/P18/z{z}/poles/P0_P18--z-{z}_optiresum-False.npy".format(z=0.38))[0]
P0_true = np.load(path_to_repo+"data/P18/z{z}/poles/P0_P18--z-{z}_optiresum-False.npy".format(z=0.38))[1]
plt.figure(figsize=(4,3))
for Vi in [1000, 2000, 3000, 4000, 5000]:
    cov = np.load(path_to_repo+"data/P18/z{z}/covs/cov_P18--z-{z}_Vs-{V}.npy".format(z=0.38, V=Vi))
    plt.plot(klin, np.sqrt(cov.diagonal())[:len(klin)]/P0_true, label=str(Vi))
plt.axhline(0.01, color='grey', linestyle='--')
plt.legend(title=r'$V_s^{1/3} \ [\mathrm{Mpc}\ h^{-1}]$', ncol=2, fontsize=legend_font)
plt.yscale('log')
plt.xlabel(r'$k \ [h \ \mathrm{Mpc}^{-1}]$', fontsize=label_font)
plt.ylabel(r'$\sigma(k)\ / \ P_0(k)$', fontsize=label_font)
plt.xticks(fontsize=label_font)
plt.yticks(fontsize=label_font)
plt.tight_layout()
plt.savefig(path_to_repo+"plots/cov_per_err--z-0.38.png", dpi=300, facecolor='white')
```

```python
klin = np.load(path_to_repo+"data/P18/z{z}/poles/P0_P18--z-{z}_optiresum-False.npy".format(z=0.38))[0]
for zi in [0.38, 0.51, 0.61]:
    P0_true = np.load(path_to_repo+"data/P18/z{z}/poles/P0_P18--z-{z}_optiresum-False.npy".format(z=zi))[1]
    for Vi in [1000, 2000, 3000, 4000, 5000]:
        cov = np.load(path_to_repo+"data/P18/z{z}/covs/cov_P18--z-{z}_Vs-{V}.npy".format(z=zi, V=Vi))
        plt.plot(klin, np.sqrt(cov.diagonal())[:len(klin)]/P0_true, label=str(Vi))
    plt.axhline(0.01, color='grey', linestyle='--')
    plt.legend(title=r'$V_s^{1/3} \ [\mathrm{Mpc}\ h^{-1}]$', ncol=2, fontsize=legend_font)
    plt.yscale('log')
    plt.xlabel(r'$k \ [h \ \mathrm{Mpc}^{-1}]$')
    plt.ylabel(r'$\sigma(k)\ / \ P_0(k)$')
    plt.show()
```
